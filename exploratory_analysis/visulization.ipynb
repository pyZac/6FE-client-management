```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#load the data
df = pd.read_csv("payments.csv")
```

```python
#count num of users per plan
plan_counts = df["paymentPlan"].value_counts()
```

```python
#Plot a bar chart
plt.figure(figsize=(6,4))
ax = plan_counts.plot(kind="bar", color="skyblue")
#add labels and title
plt.title("number of users per paymentPlan")
plt.xlabel("payment plan")
plt.ylabel("nubmber of users")
plt.xticks(rotation=0)
```

```python
mentor_count = df["attendedBy"].value_counts().head()
```

```python
mentor_count.plot(kind="bar", color="skyblue")
plt.title("top 5 mentors (referrers")
plt.xlabel("names of mentores")
plt.ylabel("number of attendes")
```
